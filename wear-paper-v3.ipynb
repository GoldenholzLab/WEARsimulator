```python
# summarize all 100k cases for SUDEP

import pandas as pd

p1 = pd.read_csv('SUDEP_case10k-1.csv')
Slist = ['SUDEP_case10k-2.csv',
         'SUDEP_case20k-1.csv','SUDEP_case20k-2.csv',
         'SUDEP_case20k-3.csv','SUDEP_case20k-4.csv']
for fn in Slist:
    p2 = pd.read_csv(fn)
    p1['SUDEP'] += p2['SUDEP']
    p1['nearSUDEP'] += p2['nearSUDEP']
p1['%prevented'] = p1['nearSUDEP'] / (p1['nearSUDEP'] + p1['SUDEP'])
p1.to_csv('SUDEP_case100k.csv',index=False)
p1
```
