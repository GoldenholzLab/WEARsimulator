```python
from weargroup_v2 import get_a_qualified_patient
import numpy as np
from tqdm.auto import trange
import matplotlib.pyplot as plt
minSz = 4
baseline = 2
trialDur = 5
clinTF = True
sensitivity=1

PCB = 0
DRG = .4
useDRG= False
accountForFAR=True
accountForSens=False
inflater=2/3
reps = 5000

FARlist = np.array([0,.05,.2,.5,1])
mlist = np.zeros(len(FARlist))
for fi,FAR in enumerate(FARlist):
    mm = np.zeros(reps)
    for i in trange(reps):
        x = get_a_qualified_patient(minSz,baseline,trialDur,clinTF,sensitivity,FAR,PCB,DRG,useDRG,inflater,accountForFAR,accountForSens)
        mm[i] = np.mean(x)

    mlist[fi]=np.median(mm)
plt.plot(FARlist,mlist,'-x')
plt.grid(visible=True)
plt.xlabel('FAR per 24hrs')
plt.ylabel('Median rate seizures/month')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.poisson(lam=0.5,size=100)
plt.plot(x,label='memoryless')
boxcar = np.array([1,1,1])
y = np.convolve(x>1,boxcar,mode='valid') * 3

plt.plot(y,label='with memory')
plt.legend()
plt.show()
```
